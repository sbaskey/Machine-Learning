```python
import pandas as pd
```

```python
input_data = pd.read_excel('marketing_campaign.xlsx')
```

```python
input_data.describe()
```

```python
x = input_data[['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','Online','CreditCard']]
y = input_data[['Offer_acceptance']]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn import tree
```

```python
from sklearn import metrics
```

```python
ctree = tree.DecisionTreeClassifier(max_depth=5, criterion='entropy',min_samples_split=3)
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.4)
ctree_model = ctree.fit(x_train,y_train)
y_train_pred = ctree_model.predict(x_train)
print('train_accuracy:',metrics.accuracy_score(y_train,y_train_pred))
y_test_pred = ctree_model.predict(x_test)
print('test_accuracy:',metrics.accuracy_score(y_test,y_test_pred))
```

```python
comparision_df = x_test.copy()
comparision_df['actual'] = y_test
comparision_df['Prediction'] = y_test_pred
comparision_df
```

```python
comparision_df = x_train.copy()
comparision_df['actual'] = y_train
comparision_df['Prediction'] = y_train_pred
comparision_df
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(ctree_model,x,y,cv=5,scoring='accuracy')
```

```python
cross_val_score(ctree_model,x,y,cv=5,scoring='accuracy').mean()
```

```python
x_test.head()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { "criterion"       : ['gini', 'entropy']
#                "max_features"     : [4,5,6],
#                "splitter"         : ['best', 'random'],
#                "max_depth"        : [2,3],
#                "min_samples_leaf" : [2,3]
             }
```

```python
optimized_tree = GridSearchCV(ctree_model, param_grid, scoring = 'accuracy', cv=5)
```

```python
optimized_tree.fit(x_train,y_train)
```

```python
optimized_tree.best_score_
```

```python
optimized_tree.best_params_
```

```python
optimized_tree.best_estimator_
```

```python
ctree_final = optimized_tree.best_estimator_
```

```python
ctree_final
```

```python
y_pred = ctree_final.predict(x_test)
print('Accuracy', metrics.accuracy_score(y_pred, y_test))
```
